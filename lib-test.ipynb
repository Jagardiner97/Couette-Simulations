```python
def run_moose(input_file, nproc=4):
    import subprocess

    # Construct the command to run MOOSE
    cmd = [
        "conda", "run", "-n", "moose",
        "mpiexec", "-np", str(nproc),
        "./couette-opt", "-i", input_file
    ]

    # Run the command
    result = subprocess.run(cmd, capture_output=True, text=True)

    # Check if the command was successful
    if result.returncode != 0:
        print("Error running MOOSE:")
        print(result.stderr)
    else:
        print("MOOSE ran successfully:")
        print(result.stdout)

    return result
```

```python
run_moose("problems/hx/hx.i", 4)
```
